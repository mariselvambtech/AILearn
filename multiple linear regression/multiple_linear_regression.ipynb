```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#since the state column is categorical data and it is a nominal value then it should be one hot encoded(convert to numberical form) 
# and drop \the first column  to save some memory . get dummmies used to convert the categorical data to numerical form
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
#alternative way to use get dummies function
pd.get_dummies(dataset,columns=["State"],prefix_sep="-",prefix="usstate").head()

```

```python
dataset.columns
```

```python
#split the dataset to input and output variables

independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the inputs and oupt to train and test variable
X_train, X_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.3)
```

```python
#creating instance of regression 
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor
```

```python
model = regressor.fit(X_train,y_train)
```

```python
#getting the weights
model.coef_
```

```python
#getting the bias
model.intercept_
```

```python
#now get the preditions for test values
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#compare the prediciton with actual values using R square
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
#rsquare value is closer to 1..good
```

```python
#save the model
import pickle
filename= "mul_linear_reg.sav"
pickle.dump(model,open(filename,"wb"))
```
