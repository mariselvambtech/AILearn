```python
#importing pandas library
import pandas
```

```python
#read input file
dataset = pandas.read_csv("Salary_Data.csv")
```

```python
#checking the Years experice column alone
independent = dataset[["YearsExperience"]]
```

```python
#checking the salary column alone
dependent=dataset[["Salary"]]
```

```python
#split the dataset to train and test
from  sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
#try to fit in regression model algorith
from sklearn.linear_model import LinearRegression
```

```python
#creating an instance
regressor = LinearRegression()
```

```python
#fitting the training sets and making learn the model
regressor.fit(X_train,Y_train)
```

```python
#getting the weights
weights = regressor.coef_
```

```python
weights
```

```python
#getting the bias
bias = regressor.intercept_
```

```python
bias
```

```python
#predict the value for test dataset
Y_pred = regressor.predict(X_test)
```

```python
Y_pred
```

```python

#get the R square value for our predictons
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test,Y_pred)
```

```python
#rsquare score is closer to 1 and so the model learnt the inputs well
```

```python
#lets save the model now
```

```python
pickle.dump(regressor,open(filename,"wb"))
```
