#### FAISS vector stores


```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = TextLoader("movies.txt").load()  # Load the text file containing movie data
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
docs = text_splitter.split_documents(loader)  # Split the loaded documents into smaller chunks
embeddings = OllamaEmbeddings(model="llama3.1:latest")
db = FAISS.from_documents(docs, embeddings)  # Create a FAISS vector store from the documents and embeddings

query = "What is the main topic of the documents?"
docs = db.similarity_search(query)  # Perform a similarity search in the vector store
print(docs)  # Print the results of the similarity search to verify the content
```

```python
embed_vector = embeddings.embed_query("who is vijay?")  # Repeating the query to ensure consistency in results
```

```python
ans = db.similarity_search_by_vector(embed_vector)  # Perform a similarity search using the vector embeddings
print(ans)  # Print the results of the similarity search by vector to verify the content
```

```python
## save the embeddings to a file
db.save_local("faiss_index")  # Save the FAISS index to a local file
```

```python
## load the embeddings from a file
db = FAISS.load_local("faiss_index", embeddings,allow_dangerous_deserialization=True)  # Load the FAISS index from the local file
print(db)  # Print the loaded FAISS index to verify its content
```

#### Vector store as Retriever

```python
retriever = db.as_retriever()  # Convert the vector store to a retriever object for further use
print(retriever) # Print the retriever object to verify its creation
```

```python
result = retriever.invoke("What is the main topic of the documents?")  # Use the retriever to answer a question
print(result[0].page_content)  # Print the result of the retriever's query to verify the output
```

#### Chroma  DB

```python
chroma_vector_db = Chroma.from_documents(docs, embeddings)  # Create a Chroma vector store from the documents and embeddings
```

```python
answer = chroma_vector_db.similarity_search("who is vijay")  # Perform a similarity search in the Chroma vector store
print(answer)  # Print the results of the similarity search in Chroma to verify the content
```
