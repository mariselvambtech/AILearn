```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
independent= dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']].values.ravel()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from xgboost import XGBRegressor
```

```python
param_grid = {'n_estimators': [5,10,30,50,100],
              'learning_rate':[0.5,1.0,1.1],
              'booster': ['gbtree', 'gblinear', 'dart']
              }
```

```python
model = GridSearchCV(XGBRegressor(random_state=42),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
model.fit(independent,dependent)
```

```python
re = model.cv_results_
```

```python
model.best_params_
```

```python
pd.DataFrame.from_dict(re)
```
