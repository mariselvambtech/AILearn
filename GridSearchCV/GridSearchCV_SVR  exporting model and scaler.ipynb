```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.columns
```

```python
independent= dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
dependent = dataset[['charges']].values.ravel()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
from sklearn.svm import SVR
```

```python
pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('svr',SVR())
         ])
```

```python
param_grid = {'svr__kernel': ['rbf', 'sigmoid', 'linear'],
              'svr__C':[0.1,1,10,100,1000],
              'svr__gamma':['auto','scale']}
```

```python
model = GridSearchCV(estimator=pipeline,param_grid=param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
model.fit(independent,dependent)#taking long time to fit for poly
```

```python
re = model.cv_results_
```

```python
re
```

```python
model.best_params_
```

```python
pd.DataFrame.from_dict(re)
```
